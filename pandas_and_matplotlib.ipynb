### Pivot tables and graphics

We want to see how the instances are distributed between two categories: 'International plan' and 'Churn'. 

<b>Exercise 7:</b>
* Build the cross table between the features using the method <i>crosstab</i>.
* Visualize the distribution for the feature 'Churn', depending on the value of the features 'International plan', 'Voice mail plan', and 'Customer service calls'.
<b>Hint:</b> Use commands plt.sublot and sns.subplots.

Your plot should look something like this:
<img src="dist_crosstab.png">

What do you see? What conclusions can be drawn? What feature (intuitively) can be more important for Churn prediction?

```python
### Write Your code here ###

fig, axs = plt.subplots(1,3,figsize=(20,10))
df["Churn"] = df["Churn"].astype("int64") #cast Churn to Integer
vmp = pd.crosstab(df["Churn"],df["Voice mail plan"]) #creating crosstab for voice mail plan
ip = pd.crosstab(df["Churn"],df["International plan"]) #creating crosstab for international plan
csc = pd.crosstab(df["Churn"],df["Customer service calls"]) #creating crosstab for customer service calls
vmp.plot(kind="bar",ax=axs[0]) #plot the subplot in the figure
ip.plot(kind="bar",ax=axs[1]) #plot the subplot in the figure
csc.plot(kind="bar",ax=axs[2]) #plot the subplot in the figure
plt.show() #display plot
```

<b>Exercise 8:</b>

Add a new feature to the dataframe which will describe whether or not the user has done more than 3 service calls. Investigate how this feature describes the churn rate.

```python
### Write Your code here ###

service_calls = []
for i in range(0,len(df)): #iterate through dataframe rows
    if df["Customer service calls"][i] > 3: 
        service_calls.append(True) #the customer has done more than 3 service calls
    else: 
        service_calls.append(False) #the customer has not done more than 3 service calls
df["More than 3 service calls"] = service_calls  #appending the new column to the dataframe
crosstab = pd.crosstab(df["Churn"],df["More than 3 service calls"]) #creating crosstab
crosstab.plot.bar() #barplot of the crosstab
plt.show() #display plot

df["More than 3 service calls"] = df["More than 3 service calls"].astype("int64") #convert column to Integer

#mean of all numerical columns of clients that are loyal
mean_loyal = df[df["Churn"] == 0].mean(numeric_only=True)

#mean of all numerical columns of clients that are not loyal
mean_nonloyal = df[df["Churn"] == 1].mean(numeric_only=True)

print("Mean of how many loyal clients have made more than 3 service calls:", mean_loyal["More than 3 service calls"])
print("Mean of how many non-loyal clients have made more than 3 service calls:", mean_nonloyal["More than 3 service calls"])

print("\n") #spacing for readability

print("On average more clients, who have left the service, have made more than 3 service calls, than clients that did not leave the service.")
```
